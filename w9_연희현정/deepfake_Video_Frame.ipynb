```python
import cv2
import numpy as np
import face_recognition
import re

dirname = 'deepfake-data_subset1/original_videos/'
videoname = '643049'
video_path = dirname + videoname+'/'
frame_save_path = videoname+'_frame'
#'C:\Users\user\Documents\cyber17\Face_Detection\'
video = cv2.VideoCapture(video_path+videoname+'_D_002.mp4')
count = 0

videoframe = video.get(cv2.CAP_PROP_POS_FRAMES)
videocount = video.get(cv2.CAP_PROP_FRAME_COUNT)

createfolder(frame_save_path)

while video.isOpened():
    #ret     T/F 
    #frame   
    if(videoframe == videocount):
        video.open(video_path+videoname+'_D_002.mp4')
    
    #video.set(cv2.CAP_PROP_POS_MSEC, (count*1000))
    ret, frame = video.read()
    #print(frame)
    
    if ret == False:
        print("Frame doesn't Exist")
        break
    
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)    
    
    try:
        
        #face_location   ()
        face_location = face_recognition.face_locations(frame) #cnn   X. , model='cnn' (GPU    )
        top, right, bottom, left = face_location[0]
        print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

        #face_location   
        face_image = Image.fromarray(frame[top:bottom, left:right])
        #Image.fromarray(face_image).show()
        face_image.save(frame_save_path+'_D_002/%d.jpg' %count)
        print('Saved frame%d.jpg' %count)
        count += 1
        
    except :
        pass


video.release()
cv2.destroyAllWindows()
```

```python
# 
def faceDetect(path, videoname, idnum):
    eye_detect = False
    face_cascade = cv2.CascadeClassifier('haarcascade_frontface.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    info = ''
    count = 0
    
    try:
        video = cv2.VideoCapture(path)
    except:
        print('video loading failed.')
        return
    
    while True:
        ret, frame = video.read()
        if not ret:
            break
        if eye_detect:
            info = "Eye detection on"
        else:
            info = "Eye detection off"
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3,5)
        #top, right, bottom, left = faces
        print(faces)
        #for (x,y,w,h) in faces:
        print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

        #face_location   
        face_image = Image.fromarray(frame[top:bottom, left:right])
        #Image.fromarray(face_image).show()
        face_image.save(videoname+'/'+idnum+'frame %d.jpg' %count)
        print('Saved frame%d.jpg' %count)
        count += 1
        
        
```
