```python
import cv2

video = cv2.VideoCapture('sample1.mp4')
count = 0

videoframe = video.get(cv2.CAP_PROP_POS_FRAMES)
videocount = video.get(cv2.CAP_PROP_FRAME_COUNT)

while video.isOpened():
    #ret 은 프레임이 존재하지 않을때 T/F 반환
    #frame 은 프레임 반환
    if(videoframe == videocount):
        video.open('sample1.mp4')
    
    ret, frame = video.read()
    if ret == False:
        print("Frame doesn't Exist")
        
        #대기시간
        if cv2.waitKey(33) > 0:
            break
    
    try:
        cv2.imshow('sample1', frame)
        cv2.imwrite('sample1_frame/frame %d.jpg' %count, frame)
        print('Saved frame%d.jpg' %count)
        count += 1
    except :
        pass

video.release()
cv2.destropAllWindows()
```
